우리동네 연령별 인구 수 result 리스트에 저장 -> 시각화를 위해.


```python
import csv
import matplotlib.pyplot as plt

#1) 연령별 인구수 리스트 선언
result = []

# 2) open file
f=open('/content/201902_201902_연령별인구현황_월간.csv')
data = csv.reader(f)

#3) 연령별 인구수 데이터 담기
for row in data:
  if '신도림동' in row[0]:   # 신도림동을 input으로 변형시키면 사용자 입력 변수 가능/읍면동 조건
    print(row[3:]) # 0~100세 정보
   #print(len(row[3:])) # 0~100세 수
   # 자료 구조 문자열 형태. -> 리저트 리스트에 담아야함.
    for i in row[3:]:
      result.append(int(i))






# 파일닫기
f.close()

print(result)
print(len(result))



plt.title("신도림동 인구구조")
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

#1) 연령별 인구수 리스트 선언
result = []
dong = input(' 인구구조를 확인하고 싶은 동네명(동) 을 입력하세요: ')


# 2) open file
f=open('/content/201902_201902_연령별인구현황_월간.csv')
data = csv.reader(f)

#3) 연령별 인구수 데이터 담기
for row in data:
  if dong in row[0]:   # 신도림동을 input으로 변형시키면 사용자 입력 변수 가능/읍면동 조건
    print(row[3:]) # 0~100세 정보
   #print(len(row[3:])) # 0~100세 수
   # 자료 구조 문자열 형태. -> 리저트 리스트에 담아야함.
    for i in row[3:]:
      result.append(int(i. replace = (","," ")))






# 파일닫기
f.close()

print(result)
print(len(result))



plt.title(dong+ "의 인구구조")
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

# 1) 연령별 인구수 리스트 선언
result = []
dong = input('인구구조를 확인하고 싶은 동네명(동)을 입력하세요: ')

# 2) 파일 열기
f = open('/content/201902_201902_연령별인구현황_월간.csv')
data = csv.reader(f)

# 3) 연령별 인구수 데이터 담기
for row in data:
    if dong in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(",", "")))

# 파일 닫기
f.close()

print(result)
print(len(result))

plt.title(dong + "의 인구구조")
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

# 1) 연령별 인구수 리스트 선언
result = []
dong = input('인구구조를 확인하고 싶은 동네명(동)을 입력하세요: ')
dong2 = input('인구구조를 확인하고 싶은 동네명(동)을 입력하세요: ')

# 2) 파일 열기
f = open('/content/201902_201902_연령별인구현황_월간.csv')
data = csv.reader(f)

# 3) 연령별 인구수 데이터 담기
for row in data:
    if dong and dong2 in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(",", "")))

# 파일 닫기
f.close()

print(result)
print(len(result))

plt.title(dong +"과"+dong2+ "의 인구구조")
plt.style.use('ggplot')
plt.plot(result,"b")
plt.plot(result,"r")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

# 1) 연령별 인구수 리스트 선언
result1 = []
result2 = []

dong1 = input('첫 번째 인구구조를 확인하고 싶은 동네명(동)을 입력하세요: ')
dong2 = input('두 번째 인구구조를 확인하고 싶은 동네명(동)을 입력하세요: ')

# 2) 파일 열기
f = open('201902_201902_연령별인구현황_월간.csv')
data = csv.reader(f)

# 3) 연령별 인구수 데이터 담기
for row in data:
    if dong1 in row[0]:
        for i in row[3:]:
            result1.append(int(i.replace(",", "")))

    if dong2 in row[0]:
        for i in row[3:]:
            result2.append(int(i.replace(",", "")))

# 파일 닫기
f.close()
print(result1)
print(result2)
print(len(result1))
print(len(result2))

plt.title(dong1 + "과 " + dong2 + "의 인구구조")
plt.style.use('ggplot')
plt.plot(result1, label=dong1, color="b")
plt.plot(result2, label=dong2, color="r")
plt.legend()
plt.show()
```
